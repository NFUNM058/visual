## 1.抚养比与人口自然增长率的对比分析

```python
import pandas as pd
import numpy as np
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.globals import ChartType, SymbolType,ThemeType
from pyecharts.charts import Bar,Tab,Line,Map,Timeline
```

```python
df = pd.read_csv('fuyang.csv',encoding='utf8',index_col=0)
df1 = pd.read_csv('growth.csv',encoding='utf8',index_col=0)
df2 = pd.read_csv('unemp.csv',encoding='utf8',index_col=0)
df3 = pd.read_csv('consume.csv',encoding='utf8',index_col=0)
```

### （1）抚养比

```python
df
```

```python
for i in range(2011, 2019):
    print(list(zip(list(df.index),list(df["{}年".format(i)]))))
```

```python
def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2011, 2019):
        map0 = (
            Map()
            .add(
                "抚养比",list(zip(list(df.index),list(df["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}dependency ratio".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_= 20,max_=48),
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl
timeline_map().render_notebook()
```

### 现象
从2011年开始看，内蒙古、东北三省附近区域和北京上海等发达地区的抚养比特别低，而贵州和广西两地的抚养比则特别高，其他地方的抚养比都比较均衡。随着年份的增大，中国整体抚养比都随着上升。

### （2）人口自然增长率

```python
df1
```

```python
for i in range(2011, 2019):
    print(list(zip(list(df1.index),list(df1["{}年".format(i)]))))
```

```python
def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2011, 2019):
        map0 = (
            Map()
            .add(
                "人口自然增长率",list(zip(list(df1.index),list(df1["{}年".format(i)]))),"china",is_map_symbol_show = False
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title="{}race of increase".format(i),subtitle="",
                                          subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=18,font_style="italic")),
                visualmap_opts=opts.VisualMapOpts(min_=-1,max_=11.5),
            
            )
        )
        tl.add(map0,"{}年".format(i))
    return tl
timeline_map().render_notebook()
```

### 现象
从2011年开始看，东北三省的人口自然增长率特别低，西部地区的人口自然增长率整体比较高，其中新疆和西藏的人口自然增长率超过了10%，中部和东部的抚养比都比较均衡。随着年份的增大，中国整体抚养比都随着上升。

## 总结论（抚养比与人口自然增长率的对比分析）
通过两张地图对比可得，总共有两个关联点：
- 第一条关联点：两地图颜色程度分布大致相同，中部和西部大体颜色都比较深，也就是抚养比和自然增长率的占比都比较高；而沿海区域（特别是东北三省）的颜色都比较浅，说明这些地方的抚养比和自然增长率相对都比较低。
- 第二条关联点：由两张地图对比可得：随着年份的增长，中国抚养比和人口自然增长率大体增大且变化规律类似，中国东南地区的变化比较明显，由如此相似的变化规律可得中国抚养比和人口自然增长率是有一定关联的。

可猜测出抚养比与人口自然增长率呈正相关。在日常生活中就能体会到，计划生育政策的实行，使得我国的人口自然增长率逐年下降，儿童和老年人，即非劳动人口减少。而现今的劳动力人口由于暂时还没有受到计划生育的影响，数量基本保持不变。因此，抚养比会下降是显而易见的。

## 总结（失业率与抚养比的对比分析）
由两交互图对比可看出：抚养比图随着年份的增加而增加，失业率图随着年份的增加而减少，且两图变化规律十分相似（总体形状变化十分类似）：由此可推测抚养比与失业率呈负相关。随着失业率的升高，经济发展会受到一定程度的阻碍，人民生活将受到负面的影响，由于不能确定自己家庭的状况，许多人会重新考虑是否要孩子。这样一来，儿童数量会减少。而对于老年人，也会因为家里劳动人口的失业而不能有良好的赡养环境，死亡数增高，使老年人口数量降低。这样一来，儿童数量会减少。从而使非劳动力人口数量减少，抚养比降低。

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line,Scatter


def grid_vertical() -> Grid:
    bar = (
        Bar()
        .add_xaxis(['2018年','2017年', '2016年', '2015年', '2014年', '2013年', '2012年', '2011年'])
        .add_yaxis("人均消费", [52912, 48883, 39200, 32334, 32001, 30003, 28868])
        
        .set_global_opts(title_opts=opts.TitleOpts(title="Grid-Bar"))
    )
    line = (
        Line()
        .add_xaxis(['2018年','2017年', '2016年', '2015年', '2014年', '2013年', '2012年', '2011年'])
        .add_yaxis("抚养比", [21.3, 21.9, 22.7, 23, 26.2, 29.2, 30.6])
        
        
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Grid-Line", pos_top="48%"),
            legend_opts=opts.LegendOpts(pos_top="48%"),
        )
    )

    grid = (
        Grid()
        .add(bar, grid_opts=opts.GridOpts(pos_bottom="60%"))
        .add(line, grid_opts=opts.GridOpts(pos_top="60%"))
    )
    return grid
```

```python
grid_vertical().render_notebook()
```

## 总结（北京市居民消费水平与抚养比的对比分析）
- 由图分析可得，从2012年-2018年，随着北京人均消费不断的增高，北京抚养比却一直往下走，且趋势和变化幅度十分类似，可猜测出人均消费和抚养比有一定关联。

人均消费提高的直接因素就是经济的提升，因为经济的提高，人们生子相对变多，那么社会劳动力都在为老龄化服务的现象就会减少，而创造更多社会价值，生产力继而上升，从而抚养比降低。
